```python
import os
import hashlib
from tqdm import tqdm



# Function to calculate the MD5 hash of a file
def get_file_hash(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()



def verify_repeated_values(folder_of_iterations):
    # Dictionary to store image hashes and their corresponding folders
    image_dict = {}

    # Iterate over all subfolders
    for subfolder in os.listdir(folder_of_iterations):
        subfolder_path = os.path.join(folder_of_iterations, subfolder)
        
        # Check if it's a directory
        if os.path.isdir(subfolder_path):
            # Iterate over all files in the subfolder
            for filename in os.listdir(subfolder_path):
                # Check if the file is a PNG image
                if filename.endswith('.png'):
                    file_path = os.path.join(subfolder_path, filename)
                    file_hash = get_file_hash(file_path)

                    # open the file .txt with same filename
                    txt_file_path = file_path.replace('.png', '.txt')
                    with open(txt_file_path, 'r') as file:
                        data = file.read().replace('\n', '')
                    
                    # If the image hash is already in the dictionary, print a message
                    if file_hash in image_dict and image_dict[file_hash] != subfolder:
                        previous_data = image_dict[file_hash]
                        if previous_data != data:
                            print(f"Image {filename} has more than one value: {previous_data} and {data}")
                    # Otherwise, add the image hash to the dictionary
                    else:
                        image_dict[file_hash] = data



for recipe_stat in tqdm(recipe_stats_objects):
    folder_of_iterations = os.path.join(mainDirectory, recipe_stat.folder_name,recipe_stat.persona_name)
    verify_repeated_values(folder_of_iterations)
```
