```python
import pandas as pd

# Assuming df is your DataFrame
df = pd.read_csv('C:\\Users\\Eduardo\\Documents\\Projects\\AutoTelos\\new_telos\\skills.csv')

# Split the 'learns' column into a list of personas
df['learns'] = df['learns'].str.split(', ')

# Explode the 'learns' column into multiple rows
df = df.explode('learns')

# Rename the 'learns' column to 'persona'
df = df.rename(columns={'learns': 'persona'})

# Group by 'persona' and aggregate 'skill' into a list
df = df.groupby('persona')['skill'].agg(list).reset_index()

# Rename the 'skill' column to 'skills'
df = df.rename(columns={'skill': 'skills'})

#make it so all persona names are in lowercase
df['persona'] = df['persona'].str.lower()

#change the line with name "Seiryuu" to "seiryu"
df.loc[df['persona'] == 'seiryu', 'persona'] = 'seiryuu'

#change the line with "susanoo" to "susano-o"
df.loc[df['persona'] == 'susanoo', 'persona'] = 'susano-o'


#print all lines of the dataframe
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```
