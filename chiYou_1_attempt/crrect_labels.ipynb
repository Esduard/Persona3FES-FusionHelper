```python
import os

mainDirectory = r"C:\Users\Eduardo\Pictures\Screenshots\persona3ChiYou"

# Dictionary to store the counts
content_counts = {}


# Iterate over directories
for directory in range(1016):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(5):
        curr_text = os.path.join(fullDirectory, f"cropped_{i}.txt")

        # read contents of curr_text as string
        with open(curr_text, "r") as f:
            curr_text_contents = f.read()
        
        # Count the occurrences of each content
        if curr_text_contents in content_counts:
            content_counts[curr_text_contents] += 1
        else:
            content_counts[curr_text_contents] = 1
        
# Print the counts
for content, count in content_counts.items():
    print(f" '{content}' : {count}")
```

```python
# open 'skills.csv'
import pandas as pd


file_path = "C:\\Users\\Eduardo\\Documents\\Projects\\AutoTelos\\skills\\skills.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Create a list of all the skills in the CSV file
list_skill_csv = df['skill'].tolist()

print(list_skill_csv)
```

```python
import difflib

# List of incorrect elements
incorrect_elements = content_counts

# List of correct elements
correct_elements = list_skill_csv

# Function to find closest matches
def find_closest_matches(incorrect, correct):
    closest_matches = {}
    for item in incorrect:
        # Find the closest match in the correct list; n=1 returns the top match
        match = difflib.get_close_matches(item, correct, n=1, cutoff=0.0)
        if match:
            closest_matches[item] = match[0]
        else:
            closest_matches[item] = None  # No match found
    return closest_matches

# Find and print the closest matches
closest_matches = find_closest_matches(incorrect_elements, correct_elements)
for incorrect, corrected in closest_matches.items():
    print(f"'{incorrect}' corrected to '{corrected}'")
```
