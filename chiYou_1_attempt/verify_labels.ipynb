```python
import os

mainDirectory = r"C:\Users\Eduardo\Pictures\Screenshots\persona3Telos"

# Dictionary to store the counts
content_counts = {}


# Iterate over directories
for directory in range(10000):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(8):
        curr_text = os.path.join(fullDirectory, f"cropped_{i}.txt")

        # read contents of curr_text as string
        with open(curr_text, "r") as f:
            curr_text_contents = f.read()
        
        # Count the occurrences of each content
        if curr_text_contents in content_counts:
            content_counts[curr_text_contents] += 1
        else:
            content_counts[curr_text_contents] = 1
        
# Print the counts
for content, count in content_counts.items():
    print(f" '{content}' : {count}")
```

```python
# Dictionary to store the counts
content_counts = {}

# Iterate over directories
for directory in range(10000):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(8):
        curr_text = os.path.join(fullDirectory, f"cropped_{i}.txt")

        # read contents of curr_text as string
        with open(curr_text, "r") as f:
            curr_text_contents = f.read()
        
        # Count the occurrences of each content
        if curr_text_contents not in content_counts:
            content_counts[curr_text_contents] = {j: 0 for j in range(8)}
        
        content_counts[curr_text_contents][i] += 1

# Print the counts
for content, count in content_counts.items():
    print(f" '{content}' : {count}")
```

```python
# Dictionary to store the filenames
filenames = {}

# Iterate over directories
for directory in range(10000):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(8):
        curr_text = os.path.join(fullDirectory, f"cropped_{i}.txt")

        # read contents of curr_text as string
        with open(curr_text, "r") as f:
            curr_text_contents = f.read()
        
        # Store the filename
        if curr_text_contents not in filenames:
            filenames[curr_text_contents] = {j: [] for j in range(8)}

        if filenames[curr_text_contents][i]:
            continue
        else:
            filenames[curr_text_contents][i] = curr_text

# Print the filenames
for content, files in filenames.items():
    print(f" '{content}' : {files}")
```

```python
list_unique_skill_crops = []
# print each value nested in filenames
for content, files in filenames.items():
    for i, filename in files.items():
        print(filename)
        list_unique_skill_crops.append(filename)
print(len(list_unique_skill_crops))
```

```python
# Open the file in write mode
with open('list_unique_skill_crops.txt', 'w') as f:
    # Write each element on a new line
    for item in list_unique_skill_crops:
        f.write("%s\n" % item)
```

```python
# Dictionary to store the filenames
filenames = {}

# Iterate over directories
for directory in range(10000):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(8):
        curr_text = os.path.join(fullDirectory, f"cropped_{i}.txt")

        # read contents of curr_text as string
        with open(curr_text, "r") as f:
            curr_text_contents = f.read()
        
        # Store the filename
        if curr_text_contents not in filenames:
            filenames[curr_text_contents] = {j: [] for j in range(8)}

        if len(filenames[curr_text_contents][i]) < 10:
            filenames[curr_text_contents][i].append(curr_text)

# Print the filenames
for content, files in filenames.items():
    print(f" '{content}' : {files}")
```

```python
list_10_skill_crops = []
# print each value nested in filenames
for content, files in filenames.items():
    for i, filelist in files.items():
        for filename in filelist:
            print(filename)
            list_10_skill_crops.append(filename)
print(len(list_10_skill_crops))
```

```python
# Open the file in write mode
with open('list_10_skill_crops.txt', 'w') as f:
    # Write each element on a new line
    for item in list_10_skill_crops:
        f.write("%s\n" % item)
```

```python
# Set to store the images
images = set()

# Open the file in write mode
with open('list_10_skill_crops_images.txt', 'w') as f:
    # Write each element on a new line
    for item in list_10_skill_crops:
        # Replace 'cropped_i.txt' with '.png'
        item = item.rsplit('\\', 1)[0] + '.png'
        
        # Check if the image has already been written
        if item not in images:
            f.write("%s\n" % item)
            images.add(item)
```
