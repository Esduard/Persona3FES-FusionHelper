```python
from PIL import Image
import os
import pytesseract

# Main directory
mainDirectory = r"C:\Users\Eduardo\Pictures\Screenshots\persona3Telos"

# Define the path to the custom dictionary
custom_dict_path = r"skills_dict.txt"

# Custom configuration for pytesseract
custom_config = r'--tessdata-dir "C:\Program Files\Tesseract-OCR\tessdata" --oem 3 --psm 7'

# Iterate over directories
for directory in range(10000):
    fullDirectory = os.path.join(mainDirectory, str(directory))

    # Iterate over images in each directory
    for i in range(8):
        curr_image = os.path.join(fullDirectory, f"cropped_{i}.png")

        # Check if the image file exists
        if os.path.exists(curr_image):
            img = Image.open(curr_image)
            text = pytesseract.image_to_string(img, config=custom_config)


            print(f"Text from image {i} in directory {fullDirectory}:")
            print(text)

            # Save the extracted text to a file
            with open(os.path.join(fullDirectory, f"cropped_{i}.txt"), 'w') as f:
                f.write(text)
        else:
            print(f"Image {i} in directory {directory} does not exist.")
```
