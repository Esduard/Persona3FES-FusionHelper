```python
import cv2
import pytesseract
from PIL import Image
import pandas as pd
import re
import difflib
import numpy as np

#pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # replace with your actual path
# --tessdata-dir "C:\Program Files\Tesseract-OCR\tessdata" 
custom_config = r'--psm 13'

# You might need to adjust these parameters according to your specific image
num_rows = len(list(arcana_names))  # Set the number of rows in your grid
num_cols = len(list(arcana_names))  # Set the number of columns in your grid
start_x = 87  # X coordinate for the upper-left rectangle
start_y = 32  # Y coordinate for the upper-left rectangle
rectangle_width = 80  # Width of each rectangle
rectangle_height = 19  # Height of each rectangle
image_path = "/home/eduardo/Documents/Persona3/AutoTelos/fusion_navigator/fusion_chart_reduced.jpeg"
output_csv_path = 'output_tesseract.csv'  # Path for the output CSV file
vertical_division_size_between_cells = 2
horizontal_division_size_between_cells = 2

list_arcana_names = list(arcana_names)

# Assuming list_arcana_names is your list
list_arcana_names = [name.upper() for name in list_arcana_names]

# Now all values in list_arcana_names are uppercase

all_possible_values = list_arcana_names + ['']

# Load the image
img = cv2.imread(image_path)

data = []  # List to store all the rows

dict_uniques = {}

for i in range(num_rows):
    row_data = []
    for j in range(num_cols):
        # Define the coordinates of the rectangle's corners
        x = start_x + (j * (rectangle_width + horizontal_division_size_between_cells))
        y = start_y + (i * (rectangle_height + vertical_division_size_between_cells))

        # Crop the rectangle from the image
        crop_img = img[y:y+rectangle_height, x:x+rectangle_width]

        # Tolerance for color distance from black
        tolerance = 10

        # Create a mask where all pixels that are within 10 units of black ([0, 0, 0]) are set to False (keep them black)
        mask = np.all(crop_img <= [tolerance, tolerance, tolerance], axis=-1)

        # Apply the mask to the cropped image, setting any pixel not within tolerance to white
        crop_img[~mask] = [255, 255, 255]

        # Verify if the entire image is white
        is_all_white = np.all(crop_img == [255, 255, 255])

        

        # Now convert the cropped image to a PIL Image object
        im = Image.fromarray(crop_img)
        # save the image
        im.save('/home/eduardo/Documents/Persona3/AutoTelos/fusion_navigator/temp_images/temp{}-{}.png'.format(i,j))

        if is_all_white:
            row_data.append('')
            #print(f"({i}, {j}) = all white")
            continue

        # Perform OCR on the cropped image
        text = pytesseract.image_to_string(crop_img, config=custom_config).strip('\n')
        
        # Clean the text using regex
        text = re.sub(r'[^a-zA-Z0-9 -]', '', text).strip(' -')

        # Attempt to find the closest match from a list of possible values
        match = difflib.get_close_matches(text, all_possible_values, n=3, cutoff=0.0)
        recognized_text = match[0] if match else text  # Use the matched value if found, otherwise the OCR result

        curr_value = dict_uniques.get(text, None)

        if not curr_value:
            dict_uniques[text] = recognized_text

        if curr_value != recognized_text and curr_value:
            print('ALERT: for {} -> {} != {}'.format(text,curr_value, recognized_text))

        # cases REOH -> STRENGTH
        if (text == 'REOH' and recognized_text == "STRENGTH"):
            #print(f"FIXED AEON")
            recognized_text = "AEON"

        # Append the recognized text to the row data
        row_data.append(recognized_text)

        #print(f"({i}, {j}) = {text} -> {recognized_text}")
    
    # Append the row data to the data list
    data.append(row_data)

# Create a DataFrame from the data list
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv(output_csv_path, index=False)

print(f"Data extraction complete. The DataFrame has been saved to '{output_csv_path}'.")
```
